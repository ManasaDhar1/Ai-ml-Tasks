```python
counts = cars['Doors'].value_counts().reset_index()
print(counts)
#columns=["movie","counts"]
plt.xlabel("Doors")
plt.ylabel("Counts")
plt.xticks(rotation=45,ha='right')
sns.barplot(data=counts,x="Doors",y="count",hue="Doors",palette="Set2")
```

```python
counts = cars['Gears'].value_counts().reset_index()
print(counts)
#columns=["movie","counts"]
plt.xlabel("Gears")
plt.ylabel("Counts")
plt.xticks(rotation=45,ha='right')
sns.barplot(data=counts,x="Gears",y="count",hue="Gears",palette="Set2")
```
