```python
import seaborn as sns
fig, axes = plt.subplots(2, 1, figsize=(12,10), gridspec_kw={'height_ratios': [1, 3]})

#plot the boxplot in the first (top) subplot
sns.boxplot(data=Univ["SAT"], ax=axes[0], color='skyblue', width=0.5, orient='h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("SAT")

#plot the hostogram with kde curve in the second subplot
sns.histplot(Univ["SAT"], kde=True, ax=axes[1], color="purple", bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("SAT")
axes[1].set_ylabel("Frequency")

#adjusr layout for better spacing
plt.tight_layout
plt.show()
```
